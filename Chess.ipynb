```python
import pandas as pd

# Read the CSV file
df = pd.read_csv("C:\\Users\\sidha\\OneDrive\\Sid\\Code\\csv\\chess\white_data.csv", header=None)

# Split the DataFrame into separate games based on empty lines
game_separator = df[0].str.strip().eq('')
game_data = [df.iloc[start:end] for start, end in zip(game_separator.groupby((game_separator.diff() & -1).cumsum()).cumsum(), game_separator.groupby((game_separator.diff() & 1).cumsum()).cumsum()) if not start == end]

# Initialize an empty list to store dictionaries of game data
all_games_data = []

# Iterate over each game
for game_df in game_data:
    game_info = {}
    for line in game_df[0]:
        # Split each line by the first occurrence of space
        key_value_pair = line.split(" ", 1)
        if len(key_value_pair) == 2:  # Ensure the line contains both key and value
            key, value = key_value_pair
            # Remove surrounding brackets and double quotes from the values
            value = value.strip('[]"')
            game_info[key] = value
    # Extract the desired columns and append to all_games_data
    all_games_data.append({
        'Date': game_info.get('[Date]', ''),
        'WhiteElo': game_info.get('[WhiteElo]', ''),
        'BlackElo': game_info.get('[BlackElo]', ''),
        'StartTime': game_info.get('[StartTime]', ''),
        'EndTime': game_info.get('[EndTime]', ''),
        'TimeControl': game_info.get('[TimeControl]', ''),
        'Termination': game_info.get('[Termination]', ''),
        'Result': game_info.get('[Result]', '')
    })

# Create a dataframe from the all_games_data list
final_df = pd.DataFrame(all_games_data)

# Print the dataframe
print(final_df)
```
