```python
import pandas as pd

df = pd.read_csv(r"C:\\Users\\sidha\\OneDrive\\Sid\\Code\\csv\\chess\white_data.csv", header=None)
print(df.head())
```

```python
games = df.groupby(df.index // 23).agg('\n'.join)

print(games.head())
```

```python

game_data = []

for game in games[0]:
    game_info = {}
    lines = game.split("\n")
    for line in lines:
       
        key_value_pair = line.split(" ", 1)
        if len(key_value_pair) == 2:  
            key, value = key_value_pair
            key = key.strip('[]"')
            value = value.strip('[]"')
            game_info[key] = value
    game_data.append(game_info)

final_df = pd.DataFrame(game_data)
```

```python
final_df["myElo"] = final_df["WhiteElo"]
final_df["oppElo"] = final_df["BlackElo"] 
final_df
```

```python
final_df.drop(columns = ["WhiteElo", "BlackElo", "CurrentPosition", "Timezone", "ECO", "ECOUrl", "UTCDate", "UTCTime", "Link", "1.", "Tournament"], inplace = True)
final_df.drop(columns = ["EndDate","Event", "Site", "Round", "White", "Black"], inplace = True)
```

```python
whiteGames = final_df
whiteGames
```

```python
whiteGames = whiteGames.drop(columns = "")
```

```python
whiteGames
```

```python
tc = ["600", "300", "180", "60"]
whiteGames = whiteGames.loc[whiteGames.TimeControl.isin(tc)]
whiteGames
```

```python
whiteGames["GameTime"] = pd.to_datetime(whiteGames["EndTime"])- pd.to_datetime(whiteGames["StartTime"])
whiteGames["GameTime"] = whiteGames["GameTime"].dt.seconds
whiteGames
```

```python
whiteGames.GameTime = whiteGames.GameTime.div(60)
whiteGames
```

```python
whiteGames = whiteGames.loc[whiteGames.GameTime < 22]
whiteGames
```

```python
terminationWhite = whiteGames.Termination.to_frame().copy()
import re
regex = r'\b(?:stalemate|resignation|checkmate|won on time|abandoned|repetition|agreement|timeout vs insufficient material|insufficient material)\b'
terminationWhite["Termination"] = terminationWhite["Termination"].apply(lambda x: re.search(regex, x).group())
whiteGames.Termination = terminationWhite
whiteGames
```

```python
whiteGames.set_index("Date", inplace = True)
whiteGames
```

```python
mapping = {
    "1-0" : "W",
    "0-1" : "L", 
    "1/2-1/2" : "D"
    }
whiteGames.Result = whiteGames.Result.replace(mapping)
```

```python
whiteGames["Pieces"] = "white"
blackGames["Pieces"] = "black"
```
