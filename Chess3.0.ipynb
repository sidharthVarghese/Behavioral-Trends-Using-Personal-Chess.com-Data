```python
q60 = chess['timeGap'].quantile(0.6)

# Define a function to assign sitting# based on timeGap value
def assign_sitting(row):
    if row['timeGap'] > q60:
        assign_sitting.counter += 1
    return assign_sitting.counter

# Initialize counter
assign_sitting.counter = 0

# Apply the function to create the sitting# column
chess['sitting#'] = chess.apply(assign_sitting, axis=1)
chess
```

```python
chess["sitting#"] = chess["sitting#"] + 1
chess
```

```python
chess["sitting#"] = chess["sitting#"].shift(1)
chess.drop(chess.index[-1], inplace = True)

chess
```

```python
chess['sitting#'].fillna(1, inplace=True)
chess['sitting#'] = chess['sitting#'].astype(int)

chess
```
